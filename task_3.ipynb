## Задание 3
## В игре Plants & Gardens каждый месяц проводятся тематические события, ограниченные по времени. В них игроки могут получить уникальные предметы для сада и персонажей, дополнительные монеты или бонусы. Для получения награды требуется пройти ряд уровней за определенное время. С помощью каких метрик можно оценить результаты последнего прошедшего события?

### Метрики аудитории:

1) Дневная аудитория (DAU) — количество уникальных пользователей, которые зашли в приложение в течение суток. Результативность события будет измеряться ежедневным онлайном пользователей

2) stickness ratio — коэффициент того, насколько часто пользователи, в течение одного периода, "остаются" в продукте. Будем измерять изменение липкости к продукту в результате примененного события

3) Удержание клиентов (Retention rate) - метрика, которая показывает, как пользователи возвращаются в продукт (и часто свидетельствует, нравится ли он им). 

4) Churn rate (отток клиентов) - метрика, которая показывает, сколько клиентов потеряла компания за определённый период. Не будем забывать про отрицательный эффект события

5) Посчитаем какое количество новых клиентов привлекло событие и соотнесем в итоге их долю к общему количеству клиентов

6) Воронка - последовательность действий пользователей до совершения ключевого действия. Позволит отследить количество пользователей, которые прошли ряд уровней за определенное время

### Метрики прибыльности

1) Посчитаем ARPU - средний доход с привлечённого пользователя, в случае, если игра монетизована. Считаем сколько пользователи нам принесут прибыль после проведенного события
